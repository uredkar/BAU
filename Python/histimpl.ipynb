```python
df['cum_wealth'] = (1+df['SP500']).cumprod()
df['drawdown'] = df['cum_wealth'].cummax() - df['cum_wealth']
df['drawdown_pct'] = df['drawdown']/ df['cum_wealth']
max_drawdown = df['drawdown_pct'].max()
df['drawdown_negative'] = -df['drawdown_pct']
```

```python
idx_max = df['drawdown_pct'].idxmax()
start_max_drawdown = max(df[:idx_max][df['drawdown_pct'] == 0.0].index)
```

```python
df.loc[start_max_drawdown]
```

```python
idx_max = df['drawdown_pct'].idxmax()
condition_after = df[:idx_max]['drawdown_pct'] == 0.0
max_drawdown_start = max(df[:idx_max][condition_after.values].index)
condition_before = df[idx_max:]['drawdown_pct'] == 0.0
if df[idx_max:][condition_before.values].empty:
    max_drawdown_end = max(df.index)
else:
    max_drawdown_end = min(df[idx_max:][condition_before.values].index)
print(max_drawdown_start,max_drawdown_end)
```

```python
print(df['Year'].iloc[max_drawdown_start],df['Year'].iloc[max_drawdown_end])
max_dd_start_year = df['Year'].iloc[max_drawdown_start]
max_dd_end_year = df['Year'].iloc[max_drawdown_end]
```

```python
plt.figure(figsize=(30,7))

plt.plot(df['Year'], df['drawdown_negative'], color='red', label='Drawdown')
plt.xlabel('Time')
plt.ylabel('Drawdown (%)')
plt.title('Drawdown Plot (Negative Values)')

#plt.annotate(f"Max Drawdown: {max_drawdown:.2%}\nPeriod: {max_drawdown_start} to {max_drawdown_end}",
#             xy=(max_drawdown_start, -max_drawdown), xycoords='data',
#             xytext=(10, -40), textcoords='offset points',
#             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))
#ax = plt.gca()
#ax.axvspan(max_drawdown_start, max_drawdown_end, facecolor='grey', alpha=0.3)
ax = plt.gca()
ax.fill_betweenx(y=[ax.get_ylim()[0], ax.get_ylim()[1]], 
                x1=max_dd_start_year, x2=max_dd_end_year, color='grey', alpha=0.3)

plt.legend()
plt.grid(True)
plt.show()
```
