```python
from scipy.stats import skew, skewtest
skewness = skew(df['etest_p'])
_, p_value = skewtest(df['etest_p'])

# Output the results
print("Skewness coefficient:", skewness)
print("p-value:", p_value)
if p_value < 0.05:
    print("The data is significantly right-skewed.")
else:
    print("The data is not significantly right-skewed.")
```
