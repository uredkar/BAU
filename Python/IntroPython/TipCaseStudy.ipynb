```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Command to tell Python to actually display the graphs
%matplotlib inline
```

```python
df = pd.read_csv('tips.csv')
df["tip_per"] = df["tip"]/df["total_bill"]
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.catplot(data=df, x="sex",kind='count', palette='muted');
```

```python
sns.catplot(data=df, x="smoker",  hue="sex", col="time", kind='count', palette='muted');
```

# sample or reshuffle to df

```python
df.sample(frac=1)
```

```python
sns.lineplot(data=df.sample(frac=1), y='tip_per', x='order_id', hue='sex');
```

```python
sns.lineplot(data=df.sample(frac=1), y='tip_per', x='order_id', hue='smoker');
```

```python
## On average Female top more than male but difference is not large
```

```python
sns.catplot(data=df, x='sex', y='tip_per',kind='bar',errorbar='sd');
```

```python
## On average dinner or lunch show little difference between tipping
```

```python
sns.catplot(data=df, x='time', y='tip_per',kind='bar');
```

```python
## Dinner time show outliers, female smoker tend to tip more at dinner time than male
```

```python
sns.catplot(data=df, x="smoker", y="tip_per", hue="sex", col="time", kind='box', palette='muted');
```

```python
sns.catplot(data=df, x="smoker", y="tip_per", hue="sex", col="time", kind='swarm', palette='muted');
```

```python
sns.pairplot(df, vars=['size', 'total_bill', 'tip_per'],corner=True, hue='sex');
```
