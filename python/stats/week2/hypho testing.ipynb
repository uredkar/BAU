```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Example data
group1_mean = 0
group1_std = 1
group1 = norm.rvs(loc=group1_mean, scale=group1_std, size=500)

group2_mean = 1
group2_std = 1
group2 = norm.rvs(loc=group2_mean, scale=group2_std, size=500)

# Calculate test statistic and p-value using z-test
n1 = len(group1)
n2 = len(group2)
mean_diff = np.mean(group1) - np.mean(group2)
std_error = np.sqrt((np.var(group1) / n1) + (np.var(group2) / n2))
z_statistic = mean_diff / std_error
p_value = 2 * (1 - norm.cdf(np.abs(z_statistic)))  # Two-tailed test

# Plotting
alpha = 0.05  # Set the alpha value

plt.figure(figsize=(8, 6))
plt.hist(group1, alpha=0.5, label='Group 1')
plt.hist(group2, alpha=0.5, label='Group 2')
plt.axvline(x=np.mean(group1), color='blue', linestyle='--', label='Mean Group 1')
plt.axvline(x=np.mean(group2), color='orange', linestyle='--', label='Mean Group 2')

# Calculate critical value for two-tailed test (z-test)
critical_value = norm.ppf(1 - (alpha / 2))

# Plot critical value lines
plt.axvline(x=critical_value, color='purple', linestyle='-', linewidth=2,
            label=f'Critical Value (Alpha={alpha})')
plt.axvline(x=-critical_value, color='purple', linestyle='-', linewidth=2)

plt.legend()
plt.xlabel('Data')
plt.ylabel('Frequency')
plt.title('Distribution of Example Data z-test')

# Accept or reject the null hypothesis
if p_value < alpha:
    plt.text(0.05, 0.95, 'Reject Null Hypothesis', transform=plt.gca().transAxes, fontsize=12, ha='left')
else:
    plt.text(0.05, 0.95, 'Accept Null Hypothesis', transform=plt.gca().transAxes, fontsize=12, ha='left')

plt.show()
```

When comparing variances, there are multiple tests available, each with its own assumptions and conditions. Here are the commonly used tests:

F-test: The F-test compares the ratio of variances between two samples. It assumes that the populations from which the samples are drawn follow a normal distribution. The F-test is used when comparing variances between two groups or treatments.

Levene's test: Levene's test is a robust alternative to the F-test that does not rely on the assumption of normality. It compares the medians of absolute deviations from the median instead of the variances. Levene's test is used when the assumption of normality is questionable or violated.

Bartlett's test: Bartlett's test is another test for comparing variances between groups. It assumes that the populations are normally distributed. Bartlett's test is sensitive to departures from normality and can be less robust than the F-test or Levene's test.

Chi-square test for variance: The chi-square test for variance is not commonly used to compare variances between samples. It is primarily used for testing independence or association between categorical variables.
    
    
The chi-square distribution is primarily used in tests related to categorical data and frequencies. It is used for testing independence or association between categorical variables, goodness-of-fit tests, and tests involving count data.

On the other hand, the F-distribution is primarily used in tests that involve comparing variances or ratios of variances. The F-test, specifically, compares the variances of two or more samples to determine if they are significantly different.    

```python
import numpy as np
from scipy.stats import chi2_contingency

# Observed data (actual counts)
observed = np.array([[45, 30, 15, 10]])  # Row represents categories, column represents counts

# Expected data (expected counts under the null hypothesis)
expected = np.array([[25, 25, 25, 25]])  # Assuming equal distribution

# Perform chi-square test
chi2, p_value, _, _ = chi2_contingency(observed)

# Set the significance level
alpha = 0.05

# Print the results
print("Chi-square statistic:", chi2)
print("P-value:", p_value)

if p_value < alpha:
    print("Reject the null hypothesis.")
    print("The observed distribution differs significantly from the expected distribution.")
else:
    print("Fail to reject the null hypothesis.")
    print("The observed distribution does not differ significantly from the expected distribution.")
```

```python
import numpy as np
import scipy.stats as stats

# Generate two sample datasets
np.random.seed(0)
sample1 = np.random.normal(loc=10, scale=2, size=100)
sample2 = np.random.normal(loc=10, scale=2, size=100)

# Null hypothesis: No difference between the means (mean1 - mean2 = 0)
# Alternative hypothesis 1: There is a difference between the means (mean1 - mean2 != 0)
# Alternative hypothesis 2: Mean of sample 1 is greater than mean of sample 2 (mean1 - mean2 > 0)
# Alternative hypothesis 3: Mean of sample 1 is less than mean of sample 2 (mean1 - mean2 < 0)

# Perform two-sample t-test
t_stat, p_value = stats.ttest_ind(sample1, sample2)

# Set significance level (alpha)
alpha = 0.05

# Check different hypotheses based on p-value
if p_value > alpha:
    print("Fail to reject the null hypothesis: No difference between the means")
else:
    if t_stat > 0:
        print("Reject the null hypothesis: Mean of sample 1 is greater than mean of sample 2")
    else:
        print("Reject the null hypothesis: Mean of sample 1 is less than mean of sample 2")
```

```python
import numpy as np
import scipy.stats as stats

# Generate a sample dataset
np.random.seed(0)
sample = np.random.normal(loc=10, scale=2, size=100)

# Null hypothesis: Mean is less than 12
# Alternative hypothesis 1: Mean is greater than or equal to 12
# Alternative hypothesis 2: Mean is not equal to 12

# Perform one-sample t-test
t_stat, p_value = stats.ttest_1samp(sample, 12)

# Set significance level (alpha)
alpha = 0.05

# Check different hypotheses based on p-value
if p_value > alpha:
    print("Fail to reject the null hypothesis: Mean is less than 12")
else:
    if t_stat > 0:
        print("Reject the null hypothesis: Mean is greater than or equal to 12")
    else:
        print("Reject the null hypothesis: Mean is not equal to 12")
```
