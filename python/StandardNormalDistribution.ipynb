```python
mu = 1000
sigma = 200

# Create a grid of x values
x = range(500, 1500)

# Calculate the PDF values for each x value
pdf = norm.pdf(x, mu, sigma)

# Calculate the CDF values for each x value
cdf = norm.cdf(x, mu, sigma)

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Plot the PDF on the first subplot
ax1.plot(x, pdf)
ax1.set_title('Probability Density Function')

# Plot the CDF on the second subplot
ax2.plot(x, cdf)
ax2.set_title('Cumulative Distribution Function')

# Show the figure
plt.show()
```

```python
from scipy.stats import norm
def plotComparison(X, mu, sigma, setName):
    probX = norm.cdf(X, mu, sigma)
    print(f'the probability of {X} is {probX}')
    probability_less_X = round(norm.cdf(X, mu, sigma),2)
    print(
    f"The probability that a student will score {X} or less is {probX}",
    )
    probabilityX = norm.pdf(X, mu, sigma)
    print(
        f"The probability that a student will score {X} is",
        probabilityX
    )
    z = (X - mu)/sigma
    print(f'the Zscale of {X} is {z}')
    Xmin = int(mu -( 3 * sigma))
    Xmax = int(mu + (3 * sigma))
    fig, (ax1) = plt.subplots(1, figsize = (12,4))
    #Plot Physics
    x = np.linspace(Xmin, Xmax, int(mu))
    ax1.plot(x, norm.pdf(x, loc = mu, scale = sigma), color = 'b')
    ax1.set_title(f'Normal Distribution of {setName} scores {X}')
    ax1.set_xlabel(f'{setName} scores')
    ax1.set_ylabel('Probability')
    ax1.axvline(X, ymax = probX, linestyle = '--', color = 'green') #did not work the same way here
```

```python
def summarizeZ(mu,sigma,score,set_name):
 z = (score - mu)/sigma  
 print(f'The Z of {set_name} with a score of {score} is {z}')
 return z
```

```python


def plotSND(X1, X2):
    fig, ax = plt.subplots()
    x = np.linspace(-4,4,50)
    ax.plot(x, norm.pdf(x, loc = 0, scale = 1), color = 'b')
    ax.set_title('Standard Normal Distribution')
    ax.set_xlabel('Z-scores')
    ax.set_ylabel('Probability')
    ax.axvline(X1, ymax = 0.9, linestyle = '--', color = 'green')
    ax.axvline(X2, ymax = 0.9, linestyle = '--', color = 'black')
    plt.show()
```

```python
# Contributed by Anand Vatkar
#Try the code below
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm

# Creating a DataFrame
df = pd.DataFrame({
    "Subject": ["Physics", "History", "Computer Science"],
    "Scores": [56.88, 77.1, 35.55],
    "Total_Marks": [80, 100, 50],
    "Mean": [47.5, 77, 33],
    "Std_Dev": [12.3, 8.2, 7.3]
})

# Standardizing the data- Calculate Z-scores and probabilities
df['Standard_Scores'] = ((df['Scores'] - df['Mean']) / df['Std_Dev'])
df['Probabilities'] = norm.cdf(df['Standard_Scores'])

# Plotting individual subjects
fig, axs = plt.subplots(1, 3, sharey=True, figsize=(15, 5))

for i, row in df.iterrows():
    x = np.linspace(row['Mean'] - 3*row['Std_Dev'], row['Mean'] + 3*row['Std_Dev'], 100)
    y = norm.pdf(x, row['Mean'], row['Std_Dev'])

    axs[i].plot(x, y)
    axs[i].set_title(f"{row['Subject']} Distribution")
    axs[i].vlines(row['Scores'], 0, norm.pdf(row['Scores'], row['Mean'], row['Std_Dev']), colors='r')
    axs[i].text(row['Scores'], norm.pdf(row['Scores'], row['Mean'], row['Std_Dev']), 
                f"Score={row['Scores']}\nPercentile={row['Probabilities']*100:.2f}%")

plt.tight_layout()
plt.show()

# Plotting standardized scores
fig, ax = plt.subplots(figsize=(5, 5))
x = np.linspace(-3, 3, 100)
y = norm.pdf(x, 0, 1)

ax.plot(x, y, label='Standard Normal Distribution')

for i, row in df.iterrows():
    ax.vlines(row['Standard_Scores'], 0, norm.pdf(row['Standard_Scores'], 0, 1), colors='r')
    ax.text(row['Standard_Scores'], norm.pdf(row['Standard_Scores'], 0, 1), 
            f"{row['Subject']}\nScore={row['Standard_Scores']:.2f}", fontsize=7)

plt.legend()
plt.show()
```
